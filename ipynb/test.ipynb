```python

# helper to convert a dictionaries into objects with attributes
class obj:
    def __init__(self, d):
        for k, v in d.items():
            if isinstance(v, (list, tuple)):
                setattr(self, k, [obj(x) if isinstance(x, dict) else x for x in v])
            else:
                setattr(self, k, obj(v) if isinstance(v, dict) else v)
        
from random import randrange

x0 = DISK_RADIUS
x1 = 100 * VALUE_SCALE - DISK_RADIUS
        
n_disks = 10
disks = []
for i in range(n_disks):
    disks.append({
            "currentState": {"x":randrange(x0,x1),"y":randrange(x0,x1),"dx":100,"dy":100},
            "nextState": {"x":1,"y":1,"dx":100,"dy":100}
        })
    
sim = obj({
    "disks": disks,
    "winningDiskIndex": -1,
})


active_step(sim,DISK_DISK_LUT)
```

```python
%matplotlib notebook

import itertools
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot()


sc = ax.scatter([], [], s=2*np.sqrt(DISK_RADIUS), edgecolors='blue',color='red')
ax.set_xlim(0, 100 * VALUE_SCALE)
ax.set_ylim(0, 100 * VALUE_SCALE)

steps_per_frame = 10
def frames():
    for i in itertools.count():
        yield i

def animate(frame):
    for i in range(steps_per_frame):
        active_step(sim, DISK_DISK_LUT)
        
    sc.set_offsets([[d.currentState.x, d.currentState.y] for d in sim.disks])
    
    ax.set_title(f"{frame * steps_per_frame}")
    return sc,

ani = FuncAnimation(
    fig,
    animate,
    frames=frames(),         
    interval=20,
    repeat=False,
    cache_frame_data=False,  
    save_count=None,         
)

ani
```
