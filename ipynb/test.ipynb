```python

# helper to convert a dictionaries into objects with attributes
class obj:
    def __init__(self, d):
        for k, v in d.items():
            if isinstance(v, (list, tuple)):
                setattr(self, k, [obj(x) if isinstance(x, dict) else x for x in v])
            else:
                setattr(self, k, obj(v) if isinstance(v, dict) else v)
        
    
sim = obj({
    "disks": [
        {
            "currentState": {"x":3 * DISK_RADIUS,"y":1,"dx":10,"dy":10},
            "nextState": {"x":1,"y":1,"dx":1,"dy":1}
        }
    ],
    "winningDiskIndex": -1,
})

lut = DISK_DISK_LUT.lut

sim_step.active_step(sim,lut)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
plt.rcParams["animation.html"] = "jshtml"


fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot()
sc = ax.scatter([],[])
ax.set_xlim(0,100 * VALUE_SCALE)
ax.set_ylim(0, 100* VALUE_SCALE)

def animate(frame):
    ds = sim.disks[0].currentState
    sim_step.active_step(sim,lut)
    sc.set_offsets(np.column_stack([ds.x,ds.y]))
    ax.set_title(f"{frame}")
    return ax

ani = FuncAnimation(fig, animate, frames= np.arange(0, 1000), interval = 100, repeat = False)
ani
```
